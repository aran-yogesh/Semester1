```python
import random
import math
from enum import Enum

class DwarfOutcome(Enum):
    ATTACK_HITS = "The dwarf successfully attacks!"
    DODGE = "The attack was dodged."
    DRUNK = "The dwarf is feeling drunk!"
    SOBER = "The dwarf is holding their liquor."
    ATTRACTIVE = "The dwarf thinks this dwarf is attractive!"
    NOT_IMPRESSIVE = "The dwarf is not impressed."
    UNCONSCIOUS = "The dwarf has been knocked unconscious."

class Dwarf:
    DEBUG = True  

    def __init__(self):
        self.strength = random.randint(1, 100)
        self.constitution = random.randint(1, 100)
        self.agility = random.randint(1, 100)
        self.charisma = random.randint(1, 100)

    def fight(self, opponent):
        
        attack_strength = math.floor(self.strength * 1.2)

    
        defense_agility = math.floor(opponent.agility * 0.8)

        
        attack_prob = attack_strength / (attack_strength + defense_agility)

        if random.random() < attack_prob:

            unconscious_prob = attack_strength / (attack_strength + opponent.constitution)
            if random.random() < unconscious_prob:
                return DwarfOutcome.UNCONSCIOUS
            else:
                return DwarfOutcome.ATTACK_HITS
        else:
            return DwarfOutcome.DODGE

    def drink(self, drinks=1):
        
        drunk_threshold = 1 - (self.constitution / 100)
        drunk_prob = drunk_threshold * drinks

        if random.random() < drunk_prob:
            return DwarfOutcome.DRUNK
        else:
            return DwarfOutcome.SOBER

    def woo(self, other_dwarf):
        
        charisma_diff = math.fabs(self.charisma - other_dwarf.charisma)

        if charisma_diff < 20:  # Similar charisma values
            return DwarfOutcome.ATTRACTIVE
        elif self.charisma > other_dwarf.charisma:
            
            if random.random() < 0.5:
                return DwarfOutcome.ATTRACTIVE
            else:
                return DwarfOutcome.NOT_IMPRESSIVE
        else:
            return DwarfOutcome.NOT_IMPRESSIVE
```
